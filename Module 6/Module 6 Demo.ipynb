# Transformations

```python
# DataFrame alternative to Flat Map
from pyspark.sql.functions import split, explode

fav_movies = spark.createDataFrame([("Pulp Fiction",), ("Requiem for a dream",), ("A clockwork Orange",)], ["title"])
fav_movies.select(explode(split("title", " ")).alias("word")).collect()
```

```python
# Sample

data = spark.createDataFrame([(i,) for i in range(1, 50)], ["value"])
data.sample(0.3, seed=14599).collect()
```

```python
# Intersection
java_skills = spark.createDataFrame([("Tom Mahoney",), ("Alicia Whitekar",), ("Paul Jones",), ("Rodney Marsh",)], ["name"])
db_skills = spark.createDataFrame([("James Kent",), ("Paul Jones",), ("Tom Mahoney",), ("Adam Waugh",), ("Rodney Marsh",)], ["name"])
java_skills.intersect(db_skills).collect()
```

```python
# Distinct

movieList = spark.createDataFrame([("A Nous Liberte",), ("Airplane",), ("The Apartment",), ("The Apartment",)], ["title"])
movieList.distinct().collect()
```

```python
# Union

java_skills = spark.createDataFrame([("Tom Mahoney",), ("Alicia Whitekar",), ("Paul Jones",), ("Rodney Marsh",)], ["name"])
db_skills = spark.createDataFrame([("James Kent",), ("Paul Jones",), ("Tom Mahoney",), ("Adam Waugh",)], ["name"])
java_skills.union(db_skills).collect()
```

```python
# Subtract

java_skills = spark.createDataFrame([("Tom Mahoney",), ("Alicia Whitekar",), ("Paul Jones",), ("Rodney Marsh",)], ["name"])
db_skills = spark.createDataFrame([("James Kent",), ("Paul Jones",), ("Tom Mahoney",), ("Adam Waugh",)], ["name"])
db_skills.subtract(java_skills).collect()
```

# Actions

```python
from functools import reduce
names1 = ["abe", "abby", "apple"]
reduce(lambda t1, t2: t1+t2, names1)
```

```python
#Python
#GroupByKey
from pyspark.sql.functions import collect_list, count
df = spark.createDataFrame([("a", 1), ("b", 1), ("a", 1)], ["key", "value"])
df.groupBy("key").agg(count("value").alias("count")).orderBy("key").collect()

df.groupBy("key").agg(collect_list("value").alias("values")).orderBy("key").collect()
```

```python
#ReduceByKey
from pyspark.sql.functions import sum as spark_sum
df = spark.createDataFrame([("a", 1), ("b", 1), ("a", 1)], ["key", "value"])
df.groupBy("key").agg(spark_sum("value").alias("sum")).orderBy("key").collect()
```
